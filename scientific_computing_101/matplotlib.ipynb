## Making plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(range(10), range(-5, 5))
```

## Making plots - Advanced

```python
fig, ax = plt.subplots()
ax.plot(range(10), range(-5, 5))
ax.set_xlim([-1, 10])
ax.set_ylim([-6, 6])
```

## Creating subplots

```python
fig, (ax0, ax1) = plt.subplots(1, 2)

ax0.plot(range(10), range(-5, 5))
ax0.set_title("Plot 1")
ax0.set_xlabel("X axis")
ax0.set_ylabel("Y axis")

ax1.plot(range(10), range(5, -5, -1))
ax1.set_title("Plot 2")
ax1.set_xlabel("X axis")
ax1.set_ylabel("Y axis")
```

```python
print(plt.style.available)
plt.style.use('seaborn')
```

## Making scatter plots

```python
fig, (ax0, ax1) = plt.subplots(1, 2)

ax0.plot(range(10), range(-5, 5))
ax0.set_title("Line Plot")

ax1.scatter(range(10), range(5, -5, -1))
ax1.set_title("Scatter Plot")
```

## Plotting histograms

```python
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.hist(np.random.randint(-100, 100, 10000));
```
